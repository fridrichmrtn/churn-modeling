```python
import numpy as np
import matplotlib.pyplot as plt

expected_value = 0.66

alphas = np.linspace(0.00001,100,50)
betas = alphas*(1-expected_value)/expected_value

for alpha, beta in zip(alphas, betas):
    fig, ax = plt.subplots()
    plt.hist(np.random.beta(alpha, beta, size=1000))
    plt.title("alpha:{:.2f} beta:{:.2f}".format(alpha, beta))
```
